```python
import pandas as pd
import numpy as np
```

```python
clusters = pd.read_csv('clusters.csv')
flows = pd.read_csv('flows.csv')
```

#### clusters

```python
clusters.head()
```

```python
clusters.columns
```

```python
clusters.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
clusters['id'] = [x.replace('POINT (' , '').replace(')' , '') for x in clusters.geometry]
```

```python
clusters['lat'] = [x.split(' ')[1] for x in clusters.id]
clusters['long'] = [x.split(' ')[0] for x in clusters.id]
```

```python
clusters
```

```python
flows.head()
```

```python
flows.columns
```

```python
flows.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
flows
```

```python
flows['origin'] = [x.replace('LINESTRING (' , '').replace(')' , '').split(",")[0] for x in flows['geometry']]
flows['dest'] = [x.replace('LINESTRING (' , '').replace(')' , '').split(",")[1] for x in flows['geometry']]
```

```python
flows
```

```python
clusters.to_csv("locations.csv")
flows.to_csv("flows.csv")
```
